# Criar um jogo da forca 

Criar um jogo da forca que pegue aleatoriamente palavras de um arquivo txt e construa o boneco com caracteres na linha de comando.

```python
import random

#Lista com as peças do quadro inicial
b = ['''

>>>>>>>>>>Forca>>>>>>>>>>

 +---+
 |   |
 ''',''' ''','''   |
''',''' ''',''' ''','''  ''',''' |
''',''' ''',''' ''','''  ''',''' |
     |
=========''']

#Lista com os membros do boneco
m = ['','O','','/','|','\ ','','/','','\ ','']

#Classe
class Forca:
#Método Construtor
    
    def __init__(self, p):
        self.p = p
#Posição dos membros do boneco no quadro
        self.a = '134579'
#Tamanho original da palavra
        self.t = len(p)
#Lista com _ no lugar de cada letra da palavra separados por espaços
        self.res = []
#Lista com cada letra da palavra separada por espaços
        self.pal = []

#método para montar o quadro inicial do jogo
    def quadro(self):
        print(b[0]+b[1]+b[2]+b[3]+b[4]+b[5]+b[6]+b[7]+b[8]+b[9]+b[10])
        
        r = ''
        for pl in range(0,self.t):            
            self.res.append('_')
            self.res.append(' ')
            self.pal.append(self.p[pl])
            self.pal.append(' ')
            r += '_ '
        print(r)
        #print(self.res)
        #print(self.pal)

#Método para encontrar e substituir as ocorrências das letras    
    def busca(self, l):
        c = self.pal.count(l)    
        if c == 0:
            n = self.a[0]
            b[int(n)] = m[int(n)]
            self.a = self.a.replace(n, '')
        else:            
            f = -1
            for i in range(0,c):
                f = self.pal.index(l,f+1)
                self.res[f]= l
        
#Método para montar quadros com os resultados intermediários    
    def resultado(self):
        print(b[0]+b[1]+b[2]+b[3]+b[4]+b[5]+b[6]+b[7]+b[8]+b[9]+b[10])

        r = ''
        for pl in range(0,len(self.res)):            
            r += self.res[pl]
        print(r)

#Método para verificar se todas as lacunas referentes às letras foram preenchidas
    def acertou(self):
        return self.res.count('_') == 0

#Método para verificar se o jogo terminou por todas as lacunas terem sido preenchidas ou por ter sido colocado o último membro do boneco
    def fim(self):
        return self.acertou() or b[9].count('\ ') == 1
        
#Método para ler uma palavra de forma aleatória do banco de palavras
def rand_word():
    with open("palavras.txt", "rt") as e:
        bank = e.readlines()
    return bank[random.randint(0,len(bank))].strip()

#Método Main - Execução do Programa
def main():

    #.Objeto
    jogo = Forca(rand_word())
#Apresenta quadro do início do jogo
    jogo.quadro()
#Até o jogo terminar, busca letras e apresenta resultados intermediários
    while not jogo.fim():    
        jogo.busca(input('\nQual a letra? \n'))
        jogo.resultado()
#Verifica se o loop parou por acerto ou boneco completado
    if jogo.acertou():
        print('\nParabéns! Você venceu!!')
    else:
        print('\nGame over! Você perdeu.')
        print('A palavra era ' + jogo.p)

    print('\nFoi bom jogar com você! Agora vá estudar!\n')

#Executa o programa        
if __name__ == "__main__":
    main()
```
